Generador de historias interactivas

```python
personajes =  ["la princesa" , "un dragon" , "un caballero" , "un mago" ]
lugares =  ["en un castillo" , " en el bosque" , "en la casa de Maluma" ]
acciones =  ["lucha contra" , "habla con" , "baila con" , "huye de" , " huye con" ]
```

```python
import random

#Seleccionar elementos al azar en cada lista
personaje = random.choice(personajes)
lugar = random.choice(lugares)
accion = random.choice(acciones)

#Creamos historia
historia = f"{personaje} {accion} el Chavo del 8 {lugar}."

#imprime la historia generada
print(historia)
```

```python
if personaje == "la princesa" : print("¡Es hora de una historia de princesas!")
else : print("¡Esta historia será de aventuras!")
```

```python
#Repetir el proceso para generar multiples historias
#Sistaxis básica de un bucle for es: for variable in secuencia (lista, rango o numeros)
import random
for i in range(10):
  personaje = random.choice(personajes)
  lugar = random.choice(lugares)
  accion = random.choice(acciones)

#Crear la historia
  historia = f"{personaje} {accion} {lugar}."
  print(f"historia {i+1}: {historia} ")
```
